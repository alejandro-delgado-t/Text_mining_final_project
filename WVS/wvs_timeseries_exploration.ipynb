## Clean dataset with non-relevant answers and aggregating by survey wave using weights
Remember the interepretation of weights: https://www.worldvaluessurvey.org/WVSContents.jsp?CMSID=WEIGHT&CMSID=WEIGHT. Below, we will use the weight `S017`, which is a weighting method based on "each participant [of a] country" (instead of `S018` which is intended for cross-country comparisons).

```python
df = pd.read_csv('US_WVS_timeseries_relevant.csv')

df.head()
```

Legend for the variables:
1. **Work and Economy**: C006: How satisﬁed are you with the ﬁnancial situation of your household? (from 1 - dissatisfied to 10 - satisfied).
2. **Attitude towards police**: E069_06: Conﬁdence towards the police (from 1 - extreme confidence to 4 - no confidence at all).
3. **Politics and democracy**: E069_07: Conﬁdence towards the parliament; E069_17: Confidence towards the justice system/courts (from 1 - extreme confidence to 4 - no confidence at all).
4. **Pride of nationality**: G006. How proud are you to be of nationality of this country? (from 1 - very proud to 4 - not proud at all).
5. **Health**:  A009: State of health (subjective) (from 1 - very good to 5 - very poor).

Now, we will invert the values for the all of the columns of interest except for `C006` (since the interpretation of those is antiintuitive - lower values are better) and normalize *all* the columns to [0, 1], according to the scale of the questions of the survey. In all cases, we ignore negative values (which will be converted to NaN), which denote that answers have not been given or that the question was not asked.

```python
for column in ['E069_06', 'E069_07', 'E069_17', 'G006']:  # Columns with scale of 1 (max) to 4 (min)
    max_value = 4
    min_value = 1
    df[column] = np.where(df[column] >= 0, (max_value - df[column]) / (max_value - min_value), np.nan)  # Inverse normalization

for column in ['A009']:  # Columns with scale of 1 (max) to 5 (min)
    max_value = 5
    min_value = 1
    df[column] = np.where(df[column] >= 0, (max_value - df[column]) / (max_value - min_value), np.nan)  # Inverse normalization

for column in ['C006']:  # Columns that only have to be normalized, with scale of 1 (min) to 10 (max)
    max_value = 10
    min_value = 1
    df[column] = np.where(df[column] >= 0, (df[column] - min_value) / (max_value - min_value), np.nan)  # Regular normalization

df.head()
```

```python
# Define columns and their respective scales
columns_info = {
    'E069_06': {'scale': (1, 4), 'normalize': 'inverse'},  # Column with scale of 1 (max) to 4 (min)
    'E069_07': {'scale': (1, 4), 'normalize': 'inverse'},  # Column with scale of 1 (max) to 4 (min)
    'E069_17': {'scale': (1, 4), 'normalize': 'inverse'},  # Column with scale of 1 (max) to 4 (min)
    'G006': {'scale': (1, 4), 'normalize': 'inverse'},  # Column with scale of 1 (max) to 4 (min)
    'A009': {'scale': (1, 5), 'normalize': 'inverse'},  # Column with scale of 1 (max) to 5 (min)
    'C006': {'scale': (1, 10), 'normalize': 'regular'}  # Column that only have to be normalized, with scale of 1 (min) to 10 (max)
}

# Function to normalize a column
def normalize_column(df, column, scale, method='regular'):
    max_value, min_value = scale
    if method == 'inverse':
        # Inverse normalization: higher values become lower and vice versa, ignoring negative values
        df[column] = np.where(df[column] >= 0, (max_value - df[column]) / (max_value - min_value), np.nan)
    elif method == 'regular':
        # Regular normalization: scale between 0 and 1
        df[column] = np.where(df[column] >= 0, (df[column] - min_value) / (max_value - min_value), np.nan)
    return df

# Apply normalization to each column
for column, info in columns_info.items():
    df = normalize_column(df, column, info['scale'], info['normalize'])

# Display the first few rows of the modified DataFrame
df.head()
```

Now, the legend for the variables is all normalized to [0, 1], with 1 representing the best of each answer and 0 the worst.
1. **Work and Economy**: C006: How satisﬁed are you with the ﬁnancial situation of your household? 
2. **Attitude towards police**: E069_06: Conﬁdence towards the police 
3. **Politics and democracy**: E069_07: Conﬁdence towards the parliament; E069_17: Confidence towards the justice system/courts 
4. **Pride of nationality**: G006. How proud are you to be of nationality of this country?
5. **Health**:  A009: State of health (subjective)

```python
# Define values to remove
# values_to_remove = {-5, -4, -3, -2, -1}  # Original values
values_to_remove = {np.nan}  # Values were converted to NaN!

columns_of_interest = ['C006', 'E069_06', 'E069_07', 'E069_17', 'G006', 'A009']

# Get unique survey waves
survey_waves = sorted(df['S002VS'].unique())

# Initialize avg_values_wave with NaN for each column and wave
avg_values_wave = {'S002VS': survey_waves}
for column in columns_of_interest:
    avg_values_wave[column] = [np.nan] * len(survey_waves)

def weighted_avg(group, value_column, weight_column):  # Function to compute weighted average
    v = group[value_column]
    w = group[weight_column]
    return (v * w).sum() / w.sum()

# Loop over all columns of interest
for column in columns_of_interest:
    # Filter out rows containing the values we want to remove
    df_filtered = df[~df[column].isin(values_to_remove)]
    # Calculate the weighted average of each column per wave using 'S017' as weights
    weighted_average_column = df_filtered.groupby('S002VS', dropna=False).apply(
        weighted_avg, column, 'S017', include_groups = False).reset_index(name=f'Weighted_Avg_{column}')
    # Merge the weighted averages with the avg_values_wave dictionary
    for i, wave in enumerate(survey_waves):
        # Find the weighted average for the current wave
        avg_value = weighted_average_column.loc[weighted_average_column['S002VS'] == wave, f'Weighted_Avg_{column}']
        # If the wave is present, update the avg_values_wave dictionary
        if not avg_value.empty:
            avg_values_wave[column][i] = avg_value.values[0]

# Convert the dictionary to a DataFrame
df_avg_values = pd.DataFrame(avg_values_wave)

df_avg_values
```

```python
# Add columnn with years of survey wave
period_mapping = {
    1: '1981-1984',
    2: '1989-1993',
    3: '1994-1998',
    4: '1999-2004',
    5: '2005-2009',
    6: '2010-2014',
    7: '2017-2022'
}

df_avg_values['period'] = df_avg_values['S002VS'].map(period_mapping)

df_avg_values
```

Note how the value is missing for wave 4, variable E069_17 (for which this question was not asked)!

```python
# Rename variables for clarity
rename_map = {
    'S002VS': 'survey_wave_number',
    'C006': 'financial_situation', 
    'E069_06': 'confidence_police', 
    'E069_07': 'confidence_parliament', 
    'E069_17': 'confidence_courts',
    'G006': 'pride_nationality',
    'A009': 'state_health'
}

df_avg_values = df_avg_values.rename(columns = rename_map)

df_avg_values
```
