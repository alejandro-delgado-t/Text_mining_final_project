Note how there is no data for the first 2 waves. This is solved below by merging the latest version of the dataset (above) with a previous version, which includes the data for the US.

## Missing US data for the first 2 waves

```python
df2020 = pd.read_csv('../WVS_TimeSeries_1981_2020_ascii_v2_0.csv')

df2020
```

```python
df_first2waves = df2020[(df2020['S002VS'] == 1) | (df2020['S002VS'] == 2)]

for column in ['S002VS', 'S003', 'COUNTRY_ALPHA']:
    print(df_first2waves[column].value_counts())
```

This includes data for the US! Now we will only keep those observations of the US, to concatenate afterwards the datasets.

```python
df_us_missing = df_first2waves[df_first2waves['COUNTRY_ALPHA'] == 'USA']

df_us_missing
```

Check the columns that are different in the 5.0 dataset and the 2.0 dataset.

```python
diff_columns = set(df_us.columns).symmetric_difference(df_us_missing.columns)
print(len(diff_columns))
print(diff_columns)
```

Maybe all of these different columns should be dropped. For now, though, we will concatenate the two datasets.

## Merging the data

```python
df_us_concat = pd.concat([df_us, df_us_missing])

df_us_concat
```

Now, all of the waves are complete for the US!

Useful variables:
- `S002VS`: survey wave. 1 1981-1984; 2 1989-1993; 3 1994-1998; 4 1999-2004; 5 2005-2009; 6 2010-2014; 7 2017-2022
- `S003`: country code ISO 3166 (840 for the US).
- `S007`: unified respondent number. Therefore, **each row in the dataset is a different respondent!**
- `S012`: Date of interview YYYYMMDD.
- Weight variables that should be used for aggregating the data by wave (read more about it [here](https://www.worldvaluessurvey.org/WVSContents.jsp?CMSID=WEIGHT&CMSID=WEIGHT)):
  - `S017`: Weight assigned to the respondent. Weight by country.
  - `S018`: Equilibrated weight-1000. According to the reference above, we should use this to consistently aggregate the data per wave and year! In summary, "Each respondent has a weight to ensure national representativeness. Always calculate weighted means or proportions so that your country-level aggregates reflect the population accurately."
- `S020`: Year of survey YYYY.

Useful questions:
- **A (society and personal view)**: A001-A006 (importance in life of various topics), A008 (feeling of happiness), A013 (feeling of loneliness), A046-A049 (abortion), A046 (politics discussion with friends), A124_01-61 (preferences, for neighbors), A170 (overall satisfaction with life), A200-A202 (position of people of different ages in society), A208 (respect towards older people in society).
- **B (environment)**: B001-B004, B007 (can be used as proxies for the willingness to take action to tackle climate change and have better sus with the environment), B021-B023 (concern about climate change, biodiversity loss and pollution).****
- **C (employment)**: C006 (satisfaction with financial situation of HH), C033 (job satisfaction). Continue with C040.
